```python
groups = ["group1", "group2", "group3"]
values = [3, 35, 355]


plt.figure(figsize=(9, 3))
# Subplot 1: Bar chart
plt.subplot(1, 3, 1)
plt.bar(groups, values)
plt.title("Bar")

# Subplot 2: Scatter plot
plt.subplot(1, 3, 2)
plt.scatter(groups, values)
plt.title("Scatter")

# Subplot 3: Line plot
plt.subplot(1, 3, 3)
plt.plot(groups, values)
plt.title("Line")
plt.show()
```

```python
# Sample data
groups = ['Group 1', 'Group 2', 'Group 3']
values = [10, 30, 20]
x = [1, 2, 3]
y = [3, 5, 2]
errors = [1, 2, 1]
scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 3, 5, 1, 4]
hist_data = [1, 2, 1, 3, 2, 4, 3, 2, 1, 3, 4, 2, 1, 3]

# Create figure with subplots
fig, axis = plt.subplots(2, 3, figsize=(15, 8))

# 1. Bar Chart
axis[0, 0].bar(groups, values, color='blue')
axis[0, 0].set_title("Bar Chart")

# 2. Line Chart
axis[0, 1].plot(x, y, color='green', marker='o')
axis[0, 1].set_title("Line Chart")

# 3. Scatter Plot
axis[0, 2].scatter(scatter_x, scatter_y, color='purple')
axis[0, 2].set_title("Scatter Plot")

# 4. Pie Chart
axis[1, 0].pie(values, labels=groups, autopct='%1.1f%%')
axis[1, 0].set_title("Pie Chart")

# 5. Histogram
axis[1, 1].hist(hist_data, bins=5, color='orange')
axis[1, 1].set_title("Histogram")

# 6. Error Bar Plot
axis[1, 2].errorbar(x, y, yerr=errors, fmt='o', capsize=5)
axis[1, 2].set_title("Error Bar Plot")

plt.tight_layout()
plt.show()
```
