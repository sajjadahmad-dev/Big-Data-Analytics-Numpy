Given arr = np.array([8, 2, 5, 1, 7]), sort the array in ascending order.

Concatenate two arrays: np.array([1, 2, 3]) and np.array([4, 5, 6]).


```python
arr = np.array([8, 2, 5, 1, 7])
print(np.sort(arr , kind="quicksort"))
m=np.array([1, 2, 3]) 
n=np.array([4, 5, 6])
print(np.concatenate((m, n), axis=0)) 
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Step 1: Load the Iris dataset
X, y = load_iris(return_X_y=True)

# Step 2: Split the data (60% training, 40% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Step 3: Create and train the Decision Tree model
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
# Step 4: mdoel.Predict on both training and testing data of xT and xT
train_preds = model.predict(X_train)
test_preds = model.predict(X_test)

# Step 5: Calculate and print accuracy score of yT,xt yt,xt
train_acc = accuracy_score(y_train, train_preds)
test_acc = accuracy_score(y_test, test_preds)

print("Training Accuracy:", train_acc)
print("Testing Accuracy:", test_acc)
```
