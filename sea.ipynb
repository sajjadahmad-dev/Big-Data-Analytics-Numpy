```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Random 5x5 data
data = np.random.rand(5, 5)

# Plot heatmap
sns.heatmap(data, annot=True, cmap="coolwarm")  # Corrected "cnaps" to "cmap", and "coolware" to "coolwarm"
plt.title("Simple Heatmap")                      # Fixed title syntax
plt.show()                                       # Fixed "plt sjow" to "plt.show()"
```

boxplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset
tips = sns.load_dataset("tips")

# Create a boxen plot
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Box Plot of Total Bill by Day")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load iris dataset
iris = sns.load_dataset("iris")

# Create pairplot with hue as species
sns.pairplot(iris, hue="species")
plt.show
```

```python
sns.histplot(data=iris, x="sepal_length", kde=True)
plt.title("Histogram of Sepal Length with KDE")
plt.show()
```

 Heatmap with Correlation Matrix

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = sns.load_dataset("tips").corr(numeric_only=True)
sns.heatmap(df,annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.lineplot(x="total_bill", y="tip",data=df)
plt.title("Line Plot")
plt.show()
```

```python
plt.hist(x="total_bill", bins=10, color='skyblue')
plt.title("Histogram of Total Bills")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the built-in 'tips' dataset
tips = sns.load_dataset("tips")

# Select only the numeric columns
numeric_tips = tips.select_dtypes(include='number')

# Calculate the correlation matrix
corr = numeric_tips.corr()

# Plot the heatmap
sns.heatmap(corr, annot=True, cmap="coolwarm", center=0)

# Add a title
plt.title("Correlation Heatmap of Numeric Variables in Tips Dataset")

# Show the plot
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, inner="quartile")
plt.title("Violin Plot of Total Bill by Day")
plt.show()
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips, color=".25")
plt.title("Swarm Plot of Total Bill by Day")

```
