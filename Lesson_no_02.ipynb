```python

### Generate an empty 4 dimensional array fill elements in it using for loop

arr_4d = np.empty((4, 4, 4, 4))
for i in range(4):
    for j in range(4):
        for k in range(4):
            for l in range(4):
                arr_4d[i, j, k, l] = (i*j*k*l)+1
                print(arr_4d)
```
